```python
# computation of fitness function 
    
objective= abs(30 - chromosome[:,0]-2*chromosome[:,1]
               -3*chromosome[:,2]-4*chromosome[:,3])
print("Fitness object :" , objective)
```

```python
# selection of fittest chromosme
fitness= 1/ (1+objective)
print("Fitness :", fitness)
```

```python
# calculating the total of fitness function 
total=fitness.sum()
print("Total :", total)
```

```python
# calculationg Probability for each chromosome
prob=fitness/ total
print("Probability :",prob)
```

```python
#Selection using Roulette Wheel and Calculating cumulative probability
cum_sum= np.cumsum(prob)
print("Cumulative Sum :",cum_sum)
```

```python
# Genertaing Random Numbers in the range 0-1
print(chromosome.shape[0]) # => n
R=np.random.random((chromosome.shape[0]))
print("Random Number : ",R)
print(R.shape[0])
```

```python
# making a new matrix of chromosome for calculation purpose
chromosome_2=np.zeros((chromosome.shape[0],4))
for i in range(R.shape[0]):
    for j in range(chromosome.shape[0]):
        if(R[i]< cum_sum[j]):
            chromosome_2[i,:]=chromosome[j,:]
            break
    
chromosome=chromosome_2
print(chromosome)
```

```python
# crossover
R=np.random.random((chromosome.shape[0]))
print("Random values :",R)
```

```python
# cross over rate 
pc=0.25
flag =R<pc
print(flag)
```

```python
# Determining the cross chromosme
cross_chromosome= chromosome[[(i==True) for i in flag]]
print("cross chromosome :",cross_chromosome)
len_cross_chromosome= len(cross_chromosome)
print(len_cross_chromosome)
```

```python
# calculating cross values
cross_value=np.random.randint(1,3,len_cross_chromosome)
print("Cross Values : ",cross_value)
```

```python
# performing cross-over


# copying the chromosme values for calculations
cpy_chromosome=np.zeros(cross_chromosome.shape)
for i in range(cross_chromosome.shape[0]):
    cpy_chromosome[i,:]=cross_chromosome[i,:]

print(cpy_chromosome)
```

```python
if(len_cross_chromosome>1):
    for i in range(len_cross_chromosome):
        c_val=cross_value[i]
        if(i==len_cross_chromosome-1):
            cross_chromosome[i,c_val:]=cpy_chromosome[0,c_val:]
        else:
            cross_chromosome[i,c_val:]=cpy_chromosome[i+1,c_val:]
    
print("Cross-Overed Chromosome :", cross_chromosome)
```

```python
i=0 # index chromosome
j=0 # 
for x in flag:
    if(x):
        chromosome[i,:]=cross_chromosome[j,:]
        j+=1
    i+=1
print("New Chromosomes :",chromosome)
```

```python
#calculating the total number of generations
a,b= chromosome.shape[0],chromosome.shape[1]
total_gen=a*b
print("Total Generations :",total_gen)
```

```python
#mutation rate => pm
pm=0.1
no_of_mutations =int(np.round(pm*total_gen))
print("No of Mutations : ",no_of_mutations)
```

```python
#calculating the Generation number
gen_num=np.random.randint(0,total_gen-1,no_of_mutations)
print("Generated Random Numbers : ",gen_num)
```

```python
# Genertaing a random number which can replace the selected chromosome to be mutated
Replacing_num = np.random.randint(0,30,no_of_mutations)
print("Numbers to be replaced : ",Replacing_num)
```

```python
for i in range(no_of_mutations):
    a=gen_num[i]
    row=a//4
    col=a%4
    chromosome[row,col]=Replacing_num[i]
    
print("Chromosome after Mutation : ",chromosome)
```

```python
import numpy as np
n=6
epoch=0
chromosome= np.random.randint(0,30,(n,4))
print("Chromosome : ",chromosome)

while(epoch<200):
    objective= abs(30- chromosome[:,0]- chromosome[:,1]*2 -chromosome[:,2]*3-chromosome[:,3] )
    print("Fitness Object :",objective)

    fitness=1 / ( 1+ objective)
    print("Fitness : ",fitness)

    total= fitness.sum()
    print("Total ",total)

    prob=fitness/total
    print("Probability : ",prob)

    cumsum=np.cumsum(prob)
    print("Cumulative Probability : ",cumsum)

    R=np.random.random((chromosome.shape[0]))
    print("Random Numbers : ",R)

    temp_chromosome=np.zeros((chromosome.shape))

    for i in range(R.shape[0]):
        for j in range(chromosome.shape[0]):
            if(R[i] < cumsum[j]):
                temp_chromosome[i,:]=chromosome[j,:]
                break
    print(temp_chromosome)
    
    chromosome=temp_chromosome
    print("Updated chromosome : ",chromosome)
    
    R=np.random.random((chromosome.shape[0]))
    pc=0.25 
    flag= R< pc
    print(flag)
    
    cross_chromosome= chromosome[[(i==True) for i in flag]]
    print("cross chromosome :",cross_chromosome)
    len_cross_chromosome= len(cross_chromosome)
    print(len_cross_chromosome)
    
    # calculating cross values
    cross_value=np.random.randint(1,3,len_cross_chromosome)
    print("Cross Values : ",cross_value)
    
    # performing cross-over


    # copying the chromosme values for calculations
    temp_chromosome=np.zeros(cross_chromosome.shape)
    for i in range(cross_chromosome.shape[0]):
        temp_chromosome[i,:]=cross_chromosome[i,:]

    print(temp_chromosome)
    
    if(len_cross_chromosome>1):
        for i in range(len_cross_chromosome):
            c_val=cross_value[i]
            if(i==len_cross_chromosome-1):
                cross_chromosome[i,c_val:]=temp_chromosome[0,c_val:]
            else:
                cross_chromosome[i,c_val:]=temp_chromosome[i+1,c_val:]
    
    print("Cross-Overed Chromosome :", cross_chromosome)
    
    i=0 # index chromosome
    j=0 # 
    for x in flag:
        if(x):
            chromosome[i,:]=cross_chromosome[j,:]
            j+=1
        i+=1
    print("New Chromosomes :",chromosome)
    
    #calculating the total number of generations
    a,b= chromosome.shape[0],chromosome.shape[1]
    total_gen=a*b
    print("Total Generations :",total_gen)
    
    #mutation rate => pm
    pm=0.1
    no_of_mutations =int(np.round(pm*total_gen))
    print("No of Mutations : ",no_of_mutations)
    
    #calculating the Generation number
    gen_num=np.random.randint(0,total_gen-1,no_of_mutations)
    print("Generated Random Numbers : ",gen_num)
    
    # Genertaing a random number which can replace the selected chromosome to be mutated
    Replacing_num = np.random.randint(0,30,no_of_mutations)
    print("Numbers to be replaced : ",Replacing_num)

    for i in range(no_of_mutations):
        a=gen_num[i]
        row=a//4
        col=a%4
        chromosome[row,col]=Replacing_num[i]
    
    print("Chromosome after Mutation : ",chromosome)
    epoch+=1

```
